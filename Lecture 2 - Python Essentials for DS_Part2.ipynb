## 1. Matlotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
siny = y.copy()
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
plt.axis([-1, 11, -2, 2]) # x y
```

```python
plt.plot(x, siny, color="green", linestyle=":", label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="-.", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() #
plt.title("Basic Trigonometric Functions")
```

```python
plt.scatter(x, siny, color="green", label="sin(x)") # 
plt.legend()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.3) # 0.3
plt.show() # 
```
