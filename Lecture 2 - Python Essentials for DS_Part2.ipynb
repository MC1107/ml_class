## 1. Matlotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
cosy = np.cos(x)
siny = y.copy()
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
```

```python
plt.plot(x, siny, color="green", linestyle=":")
plt.plot(x, cosy, color="red", linestyle="-.")
plt.axis([-1, 11, -2, 2]) # 代替上面的两个制定x y范围的语句
```

```python
plt.plot(x, siny, color="green", linestyle=":", label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="-.", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() #表示把图例画出来
plt.title("Basic Trigonometric Functions")
```

```python
plt.scatter(x, siny, color="green", label="sin(x)") # 这里的两个坐标常作为样本的两个特征来进行定点；区别于折线图的自变量和因变量
plt.legend()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.3) # 透明度为0.3
plt.show() # 标准的二维正态分布图
```

## 2. Seaborn

```python
import warnings
```

```python
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### 2.1 Distribution

```python
sns.distplot(tips['total_bill'], bins=None, hist=True, kde=True, rug=False, fit=None, 
            hist_kws=None, kde_kws=None, rug_kws=None, 
            fit_kws=None, color=None, vertical=False, 
            norm_hist=False, axlabel=None, label=None, ax=None)

# kde是高斯分布密度图，绘图在0-1之间
# hist是否画直方图
# rug在X轴上画一些分布线
# fit可以制定某个分布进行拟合
# label legend时的值
# axlabel制定横轴的说明

# 不同bill出现的频数直方图 hist=True
```

### 2.2 Stripplot

```python
sns.stripplot(x='day', y='total_bill', hue=None, data=tips, order=None, 
                      hue_order=None, jitter=True, 
                      split=False, orient=None, 
                      color=None, palette=None, size=5, 
                      edgecolor="gray", linewidth=0, 
                      ax=None)

# x: X轴数据
# y: Y轴数据
# hue: 区分不同种类数据的column name
# data: DataFrame类型数据
# jitter: 将数据分开点，防止重叠

# 每天的bill金额分布
```

```python
# 加上性别这个hue
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips, order=None, 
                      hue_order=None, jitter=True, 
                      split=False, orient=None, 
                      color=None, palette=None, size=5, 
                      edgecolor="gray", linewidth=0, 
                      ax=None)
plt.legend(loc='best')
```

### 2.3 Swarmplot

```python
# 与stripplot类似，只是数据点不会重叠 (适合小数据量)

sns.swarmplot(x='sex', y='total_bill', hue='day', data=tips)
```

### 2.4 Boxplot

```python
sns.boxplot(x='day', y='total_bill', hue=None, data=tips, order=None, 
                    hue_order=None, orient=None, 
                    color=None, palette=None, 
                    saturation=.75, width=.8, 
                    fliersize=5, linewidth=None, 
                    whis=1.5, notch=False, ax=None)

# 每天的bill金额分布箱型图
```

```python
# 和上面的stripplot一起看
sns.boxplot(x='day', y='total_bill', hue=None, data=tips, order=None, 
                    hue_order=None, orient=None, 
                    color=None, palette=None, 
                    saturation=.75, width=.8, 
                    fliersize=5, linewidth=None, 
                    whis=1.5, notch=False, ax=None)
    
sns.stripplot(x='day', y='total_bill', hue=None, data=tips, order=None, 
                 hue_order=None, jitter=True, split=False, 
                 orient=None, color=None, palette=None, 
                 size=5, edgecolor="gray", linewidth=0, 
                 ax=None)
```

### 2.5 Jointplot

```python
from scipy import stats
g = sns.jointplot(x='total_bill', y='tip',
                      data=tips, kind="reg", 
                      stat_func=stats.pearsonr, 
                      color=None, size=6, ratio=5, 
                      space=.2, dropna=True, xlim=None, 
                      ylim=None, joint_kws=None, 
                      marginal_kws=None, annot_kws=None)
# kind: 有scatter” | “reg” | “resid” | “kde” | “hex 
```

```python
g = sns.jointplot(x='total_bill', y='tip',
                      data=tips, 
                      stat_func=None, kind = 'kde',
                      color=None, size=6, ratio=5, 
                      space=.2, dropna=True, xlim=None, 
                      ylim=None, joint_kws=None, 
                      marginal_kws=None, annot_kws=None)
```

```python
g = (sns.jointplot(x='total_bill', y='tip',data=tips).plot_joint(sns.kdeplot))

# plot_joint就是在联合分布上作图
# plot_marginals就是在边缘分布上再作图
```

### 2.6 Violin

```python
sns.violinplot(x='day', y='total_bill', 
                        hue='smoker', data=tips, order=None, 
                        hue_order=None, bw="scott", 
                        cut=2, scale="area", 
                        scale_hue=True, gridsize=100, 
                        width=.8, inner="quartile", 
                        split=False, orient=None, 
                        linewidth=None, color=None, 
                        palette='muted', saturation=.75, 
                        ax=None) 

# 对称的kde图
# 中间的白点是中位数，黑色粗线对应分位数
```

###  2.7 Pointplot

```python
sns.pointplot(x='time', y='total_bill', hue='smoker', data=tips, order=None, 
                 hue_order=None, estimator=np.mean, ci=95, 
                 n_boot=1000, units=None, markers="o", 
                 linestyles="-", dodge=False, join=True, 
                 scale=1, orient=None, color=None, 
                 palette=None, ax=None, errwidth=None, 
                 capsize=None)

# estimator：点的取值是，默认是np.mean
# 午餐和晚餐的账单金额均值 不同人群
```

### 2.8 Barplot

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips, order=None, 
               hue_order=None, estimator=np.mean, ci=95, 
               n_boot=1000, units=None, orient=None, 
               color=None, palette=None, saturation=.75, 
               errcolor=".26", errwidth=None, capsize=None, 
               ax=None)

# y轴是mean value，和点图其实差不多
```

### 2.9 Countplot

```python
sns.countplot(x='day', hue='sex', data=tips)

# 每天不同性别出现的次数
```

### 2.10 Factorplot

```python
train = pd.read_csv('https://raw.githubusercontent.com/xymzhang/Machine_Learning/master/Kaggle_Projects/Titanic/train.csv')
g = sns.factorplot(x="SibSp",y="Survived",data=train,kind="bar")
g.set_ylabels("survival probability")

# 不同SibSp的人，获救的概率
```
